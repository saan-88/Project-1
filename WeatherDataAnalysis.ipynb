## How to Analyze DataFrames ?

### .head()
It shows the first N rows of the data (by default, N = 5)

```python
data.head()
```

### .shape
It shows the total no. of rows and no. of columns of the dataframe

```python
data.shape
```

### .index
This attributes the index of the dataframe

```python
data.index
```

### .columns
It shows the name of each column

```python
data.columns
```

### .dtypes()
It shows the data-type of each column

```python
data.dtypes
```

### .unique()
In a column, it shows the unique values. It can be applied on a single column only, not the whole dataframe

```python
data["Summary"].unique()
```

```python
data.nunique()
```

### .count
It shows the total number of non-null values in each column.

It can be applied on a single column as well as on whole dataframe.

```python
data.count()
```

### .value_counts
In a column, it shows all the unique values with their count. It can be applied on single column only.

```python
data["Summary"].value_counts()
```

### .info()
Provides basic information about the dataframe

```python
data.info()
```
